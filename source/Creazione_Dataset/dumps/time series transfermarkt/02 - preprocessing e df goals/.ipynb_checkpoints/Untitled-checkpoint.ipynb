```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
plt.rcParams['figure.figsize'] = 8, 5
plt.rcParams['figure.dpi'] = 90
```

```python
df = pd.read_csv('new_data/dfNew.csv')
```

```python
df['year'] = df.age + 2018 - df.age2
```

```python
df.head(20)
```

```python
sns.boxplot(x='role', y='y', data=df)
```

```python
ns = ['Cristiano Ronaldo', 'Lionel Messi', 'Neymar', 'Gareth Bale', 'Luis Surez']

for n in ns:
    plt.plot(df[df.name==n]['year'],df[df.name==n]['y'], '.-', label=n)
plt.legend()
```
