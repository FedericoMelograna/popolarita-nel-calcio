## Instagram

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
from numpy import log
```

```python
df = pd.read_csv('new_data/top100_urls.csv')
df.head()
```

```python
df['insta_followers'] = np.nan

i=0
for url in df.insta_url:
    
    try:
        followers=int(re.findall(r'"edge_followed_by":{"count":+(\d*)', requests.get(url).text)[0])
        df.iloc[i,-1]=followers
    except: 
        pass
        
    i+=1
```

```python
df.head()
```

```python
plt.plot(log(df.insta_followers), log(df.y), '.')
```

## Facebook

```python
df['facebook_followers'] = np.nan

i=0
for url in df.face_url:
    
    try:
        followers = int(re.findall(r'<div>Piace\sa\s+([\d|.]*)', requests.get(url).text)[0].replace('.', ''))
        df.iloc[i,-1]=followers
    except: 
        pass
        
    i+=1
    print(i)
```

```python
plt.plot(log(df.insta_followers), log(df.facebook_followers), '.')
```
