### Nuovo dataset senza valori mancanti

```python
N = len(list(range(16,42)))
list_ids = np.array(list(set(df.id1))).repeat(N)
list_ages = list(range(16,42))*len(list(set(df.id1)))
dfNew1 = pd.DataFrame({'id1': list_ids,'age': list_ages})

# variabili 'costanti'
list_columns = ['name', 'foot', 'league', 'nation', 'height', 'role', 'id2']
for column in list_columns:
    df_help = df.groupby(['id1',column])['y'].mean().reset_index(drop=False).loc[:,['id1',column]]
    df_help.index = df_help.id1; df_help.drop(['id1'], axis=1, inplace=True)
    dfNew1 = dfNew1.join(df_help, on='id1')

# variabile 'ruolo' preprocessata
dfNew1.role = [x.split(' - ')[0] for x in dfNew1.role.values]    
    
# variabile età massima
df_agemax = pd.DataFrame(df.groupby(['id1'])['age'].max())
dfNew1 = dfNew1.join(df_agemax, on=['id1'], rsuffix='_max')

# variabile 'year'
dfNew1['year']=2018-dfNew1['age_max']+dfNew1['age']

# variabile y
df_y = pd.DataFrame(df.groupby(['id1', 'age'])['y'].max())
dfNew1 = dfNew1.join(df_y, on=['id1','age'])

## tolgo valori dove l'età è maggiore del massimo
dfNew1.loc[(dfNew1.y.isnull()) & (dfNew1.age<=dfNew1.age_max), 'y'] = 0
dfNew1 = dfNew1[-dfNew1.y.isnull()]

# interpolo valori mancanti
dfNew1.loc[dfNew1.y==0, 'y'] = np.nan
dfNew1.loc[dfNew1.age==16, 'y'] = 0
dfNew1.y = dfNew1.y.interpolate().round()

# tolgo righe con età = 16
dfNew1 = dfNew1[dfNew1.age!=16]

dfNew1 = dfNew1.sort_values('id1').reset_index(drop=True)
```
